```python
df = pd.read_csv('heights.csv')
df.head(2)
```

```python
sd.histplot(df.height,kde=True)
```

```python
df.height.describe()
```

```python
min_deviation,max_deviation = df.height.mean() - df.height.std(),df.height.mean() + df.height.std()
min_deviation,max_deviation
```

```python
ohne_outlier = df[(df.height > min_deviation) & (df.height < max_deviation)]
```

```python
sd.histplot(ohne_outlier.height,kde=True);
```

```python
df['zscore'] = (df.height-df.height.mean())/df.height.std()
ohne_outlier = df[(df.zscore < -3) | (df.zscore > 3)]
ohne_outlier
```
