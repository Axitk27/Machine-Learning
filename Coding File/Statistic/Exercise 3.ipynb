```python
df = pd.read_csv('movie_revenues.csv')
df.revenue.describe()
```

```python
sd.histplot(df.revenue,kde=True,log_scale=True)
```

```python
df['revenue_mln'] = df['revenue'].apply(lambda x: x/1000000)
df.revenue_mln.describe()
```

```python
df['z_score'] = (df.revenue_mln-df.revenue_mln.mean())/df.revenue_mln.std()
df_new_1 = df[(df.z_score > -3) & (df.z_score < 3)]
df_new_1
```

```python
df['Mad_data'] = abs(df.revenue_mln-df.revenue_mln.median())
MAD = df.Mad_data.median()
```

```python
df['mod_z_score'] = 0.6745*(df.revenue_mln-df.revenue_mln.median())/MAD
df_new_2 = df[(df.mod_z_score < 3.5)]
df_new_2.revenue_mln.describe()
```
