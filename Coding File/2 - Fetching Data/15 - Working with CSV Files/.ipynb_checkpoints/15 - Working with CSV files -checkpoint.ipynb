```python
import pandas as pd
```

```python
df = pd.read_csv("IPL Matches 2008-2020.csv")
df
```

# 2.File is in a URL

```python
import requests
from io import StringIO

url = "https://raw.githubusercontent.com/cs109/2014_data/master/countries.csv"
headers = {"User-Agent": "Mozilla/5.0 (Macintosh; Intel Mac OS X 10.14; rv:66.0) Gecko/20100101 Firefox/66.0"}
req = requests.get(url, headers=headers)
data = StringIO(req.text)

pd.read_csv(data)
```

# 3. Sep Parameter

```python
pd.read_csv('movie_titles_metadata.tsv',sep='\t',names = ['Serial No','Name','Release_year','Rating','Votes','Geners'])
```

# 4. Index_col Parameter

```python
#It will remove the first column and Make index_col as a main column 
pd.read_csv('aug_train.csv',index_col = 'enrollee_id')
```

# 5. Header Parameter

```python
pd.read_csv('test.csv',header = 1,index_col = '0')
```

# 6. Use_col parameter

```python
#Change the column which is required or not
pd.read_csv('aug_train.csv',usecols=['enrollee_id','gender','education_level'])
```

# 7. Skiprows and Column

```python
pd.read_csv('aug_train.csv',skiprows=[1,3])
#all of the Raw removed in between the list of skiprows
```

# 8. N rows: First 100 rows will be loaded

```python
pd.read_csv('aug_train.csv',nrows=100)
```

# 9 dtypes parameter - Chnage the column type of Parameter

```python

pd.read_csv('aug_train.csv',dtype={'target' : int}).info()
```

# 10. Handling Dates - Convert normal mode into the date Object

```python
pd.read_csv("IPL Matches 2008-2020.csv",parse_dates=['date']).info()
```

```python
def rename(name):
    if name == 'Royal Challengers Bangalore':
        return 'RCB'
    if name == 'Mumbai Indians':
        return 'MI'
    else:
        return name
```

# 11. Converters

```python
pd.read_csv('IPL Matches 2008-2020.csv',converters={'team1':rename})
```

# 12. na_values parameter = Change the value to Empty

```python
pd.read_csv('aug_train.csv',na_values=["Male"])
```

# 13 Loading a Huge data set in the chunks

```python
dfs = pd.read_csv('aug_train.csv',chunksize=5000)
```

```python
for chunks in dfs:
    print(chunks.shape)
```
