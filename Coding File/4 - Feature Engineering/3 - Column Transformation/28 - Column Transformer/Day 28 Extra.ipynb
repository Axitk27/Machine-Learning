```python
df = pd.read_csv('cars.csv')
```

```python
df['car_sold'] = df['selling_price'].apply(lambda x: 'Yes' if x > 50000 else 'No')
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test = train_test_split(df.drop(['car_sold'],axis=1),df['car_sold'],test_size=0.3,random_state=None)
```

```python
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder
from sklearn.preprocessing import OrdinalEncoder
from sklearn.preprocessing import LabelEncoder
```

```python
transformer = ColumnTransformer([('trans1',OrdinalEncoder(categories=[['Test Drive Car','First Owner','Second Owner','Third Owner','Fourth & Above Owner']]),['owner']),
                                 ('trans2',OneHotEncoder(sparse_output=False, drop = 'first'),['brand','fuel'])
                                 ],
                                remainder='passthrough')
```

```python
data1 = transformer.fit_transform(x_train)
data2 = transformer.fit_transform(x_test)
```
