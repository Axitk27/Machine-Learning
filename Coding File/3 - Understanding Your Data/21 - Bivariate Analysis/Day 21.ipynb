# 2. Barplot(Numerical Numerical)

```python
titanic = pd.read_csv('train.csv')
titanic
```

```python
#Finding out the average Age of the Person
sns.barplot(data = titanic, x = 'Pclass',y = 'Age', hue = titanic['Sex'])
```

# 3 = Boxplot(Numerical Categorical)

```python
sns.boxplot(data = titanic, x = 'Sex',y = 'Age',hue = titanic['Survived'])
```

# 3 = Distplot(Numerical Categorical)

```python
#Density plot and this will give you the output interm of the 

sns.distplot(titanic[titanic['Survived']==0]['Fare'],hist=False)
sns.distplot(titanic[titanic['Survived']==1]['Fare'],hist=False)
```

```python
pd.crosstab(titanic['Pclass'],titanic['Survived'])
```

# 4 = Heat Map graph(Categorical - Categorical)

```python
sns.heatmap(pd.crosstab(titanic['Pclass'],titanic['Survived']))
#Lighter number suggest that the People in that block died in a higher number
#Darker Mode show that in the opposite way
```

```python
#Why this is not wokring
titanic.groupby('Embarked').mean('Survived')['Survived']*100
```

# 6. ClusterMap(Categorical Categorical)

```python
sns.clustermap(pd.crosstab(titanic['Parch'], titanic['Survived']))
```
