### Filtering a DataFrame

```python
ipl
```

```python
# find all the final winners
ipl[ipl['MatchNumber'] == 'Final'][['WinningTeam','Season']]
```

```python
# how many super over finishes have occured
ipl[ipl['SuperOver'] == 'Y'].shape[1]
```

```python
# how many matches has csk won in kolkata
ipl[(ipl['WinningTeam'] == 'Chennai Super Kings') & (ipl['City'] == 'Kolkata')].shape[0]
```

```python
# toss winner is match winner in percentage
x = ipl[(ipl['TossWinner'] == ipl['WinningTeam'])].shape[0]
y = ipl.shape[0]
percentage = 100*x/y
percentage
```

```python
# movies with rating higher than 8 and votes>10000
movies[(movies['imdb_rating'] > 6) & (movies['imdb_votes'] > 10000)]['original_title']
```

```python
# Action movies with rating higher than 7.5

# mask1 = movies['genres'].str.split('|').apply(lambda x : 'Action' in x)
mask1 = movies['genres'].str.contains('Action')
mask2 = movies['imdb_rating'] > 7.5
movies[mask1 & mask2]
```

```python
# write a function that can return the track record of 2 teams against each other

def information(team_a,team_b,data):
    return data[((data['Team1'] == team_a) & (data['Team2'] == team_b)) | 
                ((data['Team2'] == team_a) & (data['Team1'] == team_b))]

x = information('Rajasthan Royals','Mumbai Indians',ipl)
x
```
